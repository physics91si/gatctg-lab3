```python
# Don't edit this function
def load_sample():
    """Return the entire text of Hamlet in a string."""
    with open('hamlet.txt') as f:
        sample = f.read()
    return sample
```

```python
# Edit this function. "pass" tells Python to do nothing.
def parse_sample(eltexto, words_only, sort_list):
    if words_only == True:
        eltexto = eltexto.lower()
        eltexto = eltexto.replace('-',' ')
    wordsIntext = {}
    wordsInText2 = []
    wordsInText = eltexto.split(None)
    for word in wordsInText:
        if words_only == True:
            word = word.rstrip('?!;:."')
            word = word.replace('\'','')
        wordsInText2.append(word)
    if sort_list == True:
        wordsInText2.sort()
    return wordsInText2
    
```

```python
parse_sample(load_sample(), True, False)
```

```python
def count_freq(lst):
    """Return a dict of word frequencies given a list of words.
    
    Arguments:
    -- lst: a list of strings
    
    Returns:
    -- freq_dict: a dictionary of word: frequency pairs. The keys are strings
       containing the words, and the values are integers indicating how many
       times each word appears.
    
    Example:
    >>> count_freq(['time', 'after', 'time'])
    {'time': 2, 'after': 1}
    """

    lst.sort()
    freq_dict = {}
    tempVar = lst.pop()
    while lst != []:
        a = tempVar
        freq_dict[tempVar] = (lst.count(tempVar) + 1)
        while a == tempVar and lst != []:
            a = lst.pop()
            if lst == [] and (a in freq_dict.keys()):
                pass
            elif a != tempVar:
                freq_dict[a] = (1)
        tempVar = a
        
    return freq_dict
def mean(lst):
    """Return the arithmetic mean of a list of numbers."""
    temp = []
    for i in lst:
        temp.append(i)
    import numpy
    return numpy.mean(temp)

def stdev(lst):
    """Return the standard deviation of a list of numbers."""
    temp = []
    for i in lst:
        temp.append(i)
    import numpy
    return numpy.std(temp)

def median(lst):
    """Return the median of a list of numbers."""
    temp = []
    for i in lst:
        temp.append(i)
    import numpy
    return numpy.median(temp)
```

```python
def print_stats(lst):
    """Print common statistics about the word frequencies of a list of words."""
    print("The median is %d letter(s) long" % (median(lst)))
    print("The mean is %d letter(s) long" % (mean(lst)))
    print("The standard deviation is %d letter(s) long" % (stdev(lst)))
```

```python
print_stats(count_freq(parse_sample(load_sample(), True, True)).values())
```
